#### Question 1: Which publisher has the highest total sales? Show the top 10 publisher by sales on a graph

```python
df1 = df.groupby('Publisher').sum()
#df1[df1['Publisher'] == 'Nintendo']
df1 = df1.sort_values(by='Global_Sales', ascending=False)
df1.head(10)
```

```python
df2 = df1.head(10)
df2
```

```python
#Put results on to grpah
df2.reset_index(inplace= True)
df2
```

```python
plt.bar(df2['Publisher'], df2['Global_Sales'])
plt.xticks(rotation='vertical', size=10)
plt.title('Publisher by Global Sales')
plt.xlabel('Publishers')
plt.ylabel('Global sales (millions)')
plt.grid()
plt.show()
```

#### Question 2: What is the total sales by region?

```python
data = {'Region' : ['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales'],
            'Total Sales': [df1['NA_Sales'].sum(), df1['EU_Sales'].sum(), df1['JP_Sales'].sum(), df1['Other_Sales'].sum()]}
df_region = pd.DataFrame(data, columns=['Region', 'Total Sales'])
df_region
```

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales'
sizes = df_region['Total Sales']
explode = (0, 0.1, 0, 0)

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=60)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
